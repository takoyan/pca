```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
data = pd.read_csv("univ-power.csv")
data= data.apply(lambda x: (x - x.mean()) / x.std(), axis=0)
```

```python
pca = PCA()
feature = pca.fit(data)
feature = pca.transform(data)
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8, c=list(data.iloc[:, 0]))
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
plt.plot([0] + list( np.cumsum(pca.explained_variance_ratio_)), "-o")
plt.xlabel("Number of principal components")
plt.ylabel("Cumulative contribution rate")
plt.grid()
plt.show()
```

```python
pd.DataFrame(pca.components_, columns=data.columns, index=["PC{}".format(x + 1) for x in range(len(data.columns))])
```
